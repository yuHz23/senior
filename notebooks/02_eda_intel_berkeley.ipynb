# Intel Berkeley EDA
Exploratory analysis of the Intel Berkeley sensor dataset.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.data.stream_simulator import INTEL_FEATURES
from src.data.intel_loader import FAULT_RULES

CSV_PATH = '../data/raw/intel_berkeley.csv'
```

```python
df = pd.read_csv(CSV_PATH, usecols=INTEL_FEATURES)
print(df.shape)
df.describe()
```

```python
print('Missing values:')
print(df.isnull().sum())
df = df.fillna(method='ffill').fillna(method='bfill')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, feat in zip(axes.flat, INTEL_FEATURES):
    lo, hi = FAULT_RULES[feat]
    ax.hist(df[feat].dropna(), bins=100)
    ax.axvline(lo, color='r', linestyle='--', label='fault bounds')
    ax.axvline(hi, color='r', linestyle='--')
    ax.set_title(feat)
plt.suptitle('Intel Berkeley Feature Distributions (red = fault bounds)')
plt.tight_layout()
plt.show()
```

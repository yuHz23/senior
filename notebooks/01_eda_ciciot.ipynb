# CIC IoT EDA
Exploratory analysis of the CIC IoT dataset before training.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.stream_simulator import CICIOT_FEATURES

CSV_PATH = '../data/raw/ciciot.csv'
LABEL_COL = 'label'
```

```python
df = pd.read_csv(CSV_PATH, usecols=CICIOT_FEATURES + [LABEL_COL], nrows=50_000)
print(df.shape)
df.head()
```

```python
print('Label distribution:')
print(df[LABEL_COL].value_counts())
```

```python
df[CICIOT_FEATURES].hist(bins=50, figsize=(14, 8))
plt.suptitle('CIC IoT Feature Distributions')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df.sample(2000), vars=CICIOT_FEATURES, hue=LABEL_COL, plot_kws={'alpha': 0.3})
plt.show()
```
